### Splitting DateTime into two coulmns namely CallDate and CallTime
###  CallTime is in AM/PM format so converting that into 24 hour format
1. CallDate - has the month and year of the incident
2. CallTime - has the hour of the day when the incident was reported

```python
df['CallDate'] = df['CallDateTime'].apply(lambda time: time[0:2] + '/' + time[6:10])
```

```python
df['CallTime'] = df['CallDateTime'].apply(lambda time: str((int(time[11:13]) + 12)) if time[20:22] == 'PM' else time[11:13]) 
                                                                                                                                       
```

```python
df.head()
```

```python
# # finding the count of records present in each rows
df.count()
```

```python
# Creating a Date Column with the format'%m/%d/%Y'
import datetime as dt
df['Date'] = df['CallDateTime'].apply(lambda time: dt.datetime.strptime(time[0:10],'%m/%d/%Y'))                                                                                                                     
```

```python
df['Date'].dtype
```

```python
# Shows the week day name
import datetime
df['Day_of_Week'] = df['Date'].dt.weekday_name
```

```python
#The day of the week with Monday=0, Sunday=6.
df['DayofWeek'] = df['Date'].dt.dayofweek
```

```python
df.head()
```

```python
df.drop(['Date','CallDateTime'],axis=1,inplace=True)
```

```python
df.head()
```

### For Analysis purpose we may need Day, Month and Year as seperate column

```python
# Finding the Year of 911 calls made
df['CallYear'] = df['CallDate'].apply(lambda date: str(date[3:7])).astype(int)
```

```python
# Finding the months of 911 calls made
df['CallMonth'] = df['CallDate'].apply(lambda date: date[0:2])
df['CallMonth'] = df['CallMonth'].astype(int)
```

```python
df2 = df.groupby(['Day_of_Week'])[['RecordID']].count()
df2.rename(columns = {'RecordID':'TotalCalls'}, inplace = True)
df2 = df2.reset_index() 
# df2.head(10)
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday']
mapping = {day: i for i, day in enumerate(days)}
key = df2['Day_of_Week'].map(mapping)
```

```python
df2 = df2.iloc[key.argsort()].set_index('Day_of_Week')
df2 = df2.reset_index() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig, ax = plt.subplots()
ax.scatter((np.arange(len(df2['Day_of_Week']))), df2['TotalCalls'],marker='o')
ax.set_xticks(np.arange(len(df2['Day_of_Week'])))
ax.set_xticklabels(df2['Day_of_Week'], rotation= 75)
#ax.xaxis.set_ticks(np.arange(len(df2['Day_of_Week'])))
#ax.xaxis.set_ticklabels(df2['Day_of_Week'], rotation= 'vertical')
plt.xlabel("Day_of_Week")
plt.ylabel("TotalCalls")

plt.show()
```

### Observation for Day of Week
- One can observe that as the week progress toward weekends starting with Monday, 911 calls increases with each passing day. Friday being the day with highest number of 911 calls.
- Also, as observed from earlier plot of Hour, we can infer that could during weekdays there is more people on road during late afternoon and evening time, where on Friday being the last working day of week, people remain in fun mood and parties goes on.
